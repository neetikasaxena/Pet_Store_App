# Pet_Store App

### Importing Libraries

```python
from datetime import datetime
import math
import sys
```

### Class Definitions

```python
weight_cal = []
# Parent Class Defination
class Pet():
    def __init__(self, name, dob, weight, p_name, p_address):
        self.name = name
        self.dob = dob
        self.weight = weight
        self.person = Person(p_name, p_address)
        current_date = datetime.today()
        self.delta = (current_date - self.dob).days
# Accessor
    def get_name(self):
        return self.name
    def get_dob(self):
        return self.dob
    def get_weight(self):
        return self.weight
    def get_person(self):
        return self.person
    def get_delta(self):
        return self.delta
    def get_person(self):
        return self.person
# Mutator
    def set_name(self,name):
        self.name = name
    def set_dob(self,dob):
        self.dob = dob
    def set_weight(self,weight):
        self.weight = weight
    def set_person(self,p):
        self.person = p

    def __str__(self):
        return "Pet Name: {}; Pet DOB: {}; Birth Weight: {} \nOwner Name: {}; Owner Address: {} \n".format(self.name, self.dob,self.weight,self.person.get_person_name(),self.person.get_address())
```

### Inheritance Class

```python
class Mammal(Pet):
    def __init__(self,name, dob, weight,p_name, p_address,litter_size, hasClaws):
        super().__init__(name, dob, weight, p_name, p_address)
        self.litter_size = litter_size
        self.hasClaws = hasClaws  
# Accessor
    def get_name(self):
        return self.litter_size
    def get_dob(self):
        return self.hasClaws

# Mutator
    def set_name(self,litter_size):
        self.litter_size = litter_size
    def set_dob(self,hasClaws):
        self.hasClaws = hasClaws
        
    def weight_increase(self):
        x=0
        y= self.weight
        while x<self.delta:
            x+=50 
            if x>self.delta: break
            if (x<=300):
                y = y*1.08
        return "{} weight is {} pounds and is {} days old".format(self.name,math.trunc(y),self.delta)
    def __str__(self):
        op = "Details of this Mammal pet are: \n"
        op += super().__str__()
        op += "Litter_Size: {}; HasClaws: {}\n".format(self.litter_size,self.hasClaws)
        return(op)
    
class Fish(Pet):
    def __init__(self,name, dob, weight,p_name, p_address, scale_condition, length):
        super().__init__(name, dob, weight,p_name, p_address)
        self.scale_condition = scale_condition
        self.length = length
# Accessor
    def get_name(self):
        return self.scale_condition
    def get_dob(self):
        return self.length

# Mutator
    def set_name(self,scale_condition):
        self.__scale_condition = scale_condition
    def set_dob(self,length):
        self.length = length
        
    def weight_increase(self):
        x=0
        y= self.weight
        while x<self.delta:
            x+=90 
            if x>self.delta: break
            if (x<=360):
                y = y* 1.04
        return "{} weight is {} pounds and is {} days old ".format(self.name,math.trunc(y),self.delta)
    def __str__(self):
        op = "Details of this Fish pet are: \n"
        op += super().__str__()
        op += "scale_conditoin: {}; Fish_Lengh: {}\n".format(self.scale_condition,self.length)
        return(op)
class Amphibian(Pet):
    def __init__(self,name, dob, weight,p_name, p_address, isVenomous):
        super().__init__(name, dob, weight,p_name, p_address)
        self.isVenomous = isVenomous
# Accessor
    def get_name(self):
        return self.isVenomous

# Mutator
    def set_name(self,isVenomous):
        self.isVenomous = isVenomous
        
    def weight_increase(self):
        x=0
        y= self.weight
        while x<self.delta:
            x+=120
            if x>self.delta: break
            if (x<=360):
                y = y*1.05
            elif(x>360)&(x<=600):
                y = y*1.03
        return "{} weight is {} pounds and is {} days old".format(self.name,math.trunc(y),self.delta)
    def __str__(self):
        op = "Details of this Amphibian pet are: \n"
        op += super().__str__()
        op += "IsVenomous: {}\n".format(self.isVenomous)
        return(op)
class Person():
    def __init__(self,person_name, address):
        self.person_name = person_name
        self.address = address
#Accessor
    def get_person_name(self):
        return self.person_name
    def get_address(self):
        return self.address
#mutator   
    def set_person_name(self,person_name):
        self.person_name = person_name
    def set_address(self,address):
        self.address = address
        
```

### Add_pet Function() To get input from user

```python
def add_pet():
    #pet_type checker
    while(True):
        try:
            pet_type = int(input("Type of Pet?(1-Mammal, 2-Fish, 3-Amphibian) :"))
            if pet_type not in (1,2,3):
                raise TypeError
                continue
            else:
                break
        except TypeError:
            print("Select Correct Value among 1,2,3")
        except ValueError:
            print("Enter Numerical Value")
    #pet_name  
    while(True):
        try:
            pet_name = input("Enter the pet's Name :").strip().capitalize()
            if len(pet_name)==0:
                raise TypeError
                continue
            else:
                break
        except TypeError:
            print("Please enter Name, don't leave it blank")
    #pet_dob checker
    try:
        pet_dob = datetime.strptime(input("Enter the Pet's date of birth (year,month,date): "),'%y,%m,%d')
    except ValueError:
        print("Sorry Exiting! - Enter date in correct format(Eg: 20,05,06)")
        sys.exit()
    #pet weight checker
    try:
        pet_weight = float(input("Enter the pet's Birth Weight: "))
    except ValueError:
        print("Sorry!! you Entered Non-Numeric value: Exiting")
        sys.exit()
    #pet owners detail
    while(True):
        try:
            owner_name = input("Enter the owner's Name: ").strip().capitalize()
            if len(owner_name)==0:
                raise TypeError
                continue
            else:
                break
        except TypeError:
            print("Please enter Owner's Name, don't leave it blank")
    while(True):
        try:
            owner_address = input("Enter the owner's Address: ").strip()
            if len(owner_address)==0:
                raise TypeError
                continue
            else:
                break
        except TypeError:
            print("Please enter Owner's Address, don't leave it blank")
    
    
    #mammal Details
    if pet_type == 3: #Apbhibian
        while(True):
            try:
                amphi_venomous = (input("Is the pet Venomous? Answer 'yes' or 'no': ")).lower().strip()
                if (amphi_venomous == 'yes')| (amphi_venomous == 'no'):
                    break
                else: 
                    raise TypeError
                    continue
            except TypeError:
                print("Enter either yes or no")
        obj = Amphibian(pet_name,pet_dob,pet_weight,owner_name,owner_address,amphi_venomous)
        
    elif pet_type == 2:# Fish
        while(True):
            try:
                fish_scale = input("Enter the scale condition: ").strip()
                if len(fish_scale)==0:
                    raise TypeError
                    continue
                else:
                    break
            except TypeError:
                print("Please enter Fish scale, don't leave it blank")

        while(True):
                try:
                    fish_length = float(input("Enter the length of Fish(inches)"))
                    if type(fish_length) == float:
                        break
                except ValueError:
                    print("Enter Numerical Value(Eg: 45.20)")
                    continue
        obj = Fish(pet_name,pet_dob,pet_weight,owner_name,owner_address,fish_scale,fish_length)
        
    else: #Mammal
        while(True):
                try:
                    mammal_litter = int(input("Enter the Litter Size: "))
                    if type(mammal_litter) == int:
                        break
                except ValueError:
                    print("Enter Numerical Value in integer(Eg: 45)")
                    continue
        while(True):
            try:
                mammal_hasClaws = (input("Is the pet has claws? Answer 'yes' or 'no': ")).lower().strip()
                if (mammal_hasClaws == 'yes')| (mammal_hasClaws == 'no'):
                    break
                else: 
                    raise TypeError
                    continue
            except TypeError:
                print("Enter either yes or no")
        obj = Mammal(pet_name,pet_dob,pet_weight,owner_name,owner_address,mammal_litter,mammal_hasClaws)

    weight_cal.append(obj)
    with open('petdata.dat', 'a+', newline ='') as f:
        f.write("\n")
        f.write(str(obj))
#########
def get_weight(weight_cal):
    if weight_cal == []:
        print("There are no Pet's data")
    else: 
        for i in weight_cal:
            print(i.weight_increase())
```

### Driver Function

```python
def driver():
    import os
    while(True):
        print("===Menu===\n1. To Add a pet\n2. To Print current weight of all pets\n3. To Print all pets and owner\n4. To Exit program")
        try:
            x = int(input("Your Selection: "))
            if (x<1)|(x>4):
                raise ValueError
                continue
            elif (x ==1):
                add_pet()
            elif (x==2):
                get_weight(weight_cal)
            elif (x==3):
                if os.path.isfile('petdata.dat') == False:
                    print('File is empty')
                else:
                    with open('petdata.dat') as f:
                        for line in f:
                            print(line, end='')

            elif (x==4):
                print("You choose to exit the program")
                final = (input("Are you sure(Y/N)?  ")).lower().strip()
                if (final == 'y'):
                    print('Data saved to petdata.dat')
                    print('Goodbye')
                    break
                elif (final=='n'): 
                    driver()
                else:
                    print("Enter either Y or N")
                    continue

        except ValueError:
            print("Enter Numerical Value among 1,2,3,4")
driver()
```
